Python can use **Comparison operators** to compare two statement and provie an answer as True (1) or False (0).  Python reserves the symbols >,>=,<,<=,!=,==, to denote “greater than”, “greater than or equal”, “less than”, “less than or equal”, “not equal”, and “equal”, respectively.



```python
not(1 > 0)
```

```python
not(1 < 0)
```

Python can also perform logical operations. The fundamental logical operators are **and**, **or**, and **not**.

```python
(1>0) and (2==3)
```

```python
(1>0) or (2==3)
```

```python
not (1>0) or (2==3)
```

```python
5+3 > 3+5
```

```python
import math

math.cos(math.pi/2) != math.sin(math.pi)
```

```python
math.sin(math.pi)
```

```python
math.cos(math.pi/2)
```

You can also work with True and False statements directly. For example

```python
True and True 
```

```python
True and False
```

```python
False and True
```

```python
False and False
```

```python
True or True, True or False, False or True, False or False
```
