## Plotting

One of the most important features for data analysis is the ability to visualize your data. Python + Jupyter Notebooks represent a great tool for plotting and visualizing your data. 

You will use the libraty ```matplotlib``` for everything related to plotting

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = [0, 1, 2, 3] 
y = [0, 1, 4, 9]
plt.plot(x, y)
plt.show()
````

````python
x = [0, 1, 2, 3] 
y = [0, 1, 4, 9]
plt.plot(x, y)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('My First Graph')
plt.show()
````

Plot the function 
$$
y = x^2 , -5 \leq x \leq 5
$$

````python
x = np.linspace(-5,5,100)
y = x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('$y = x^2$')
plt.show()
````

You can also change the type of line and color.



````python
x = np.linspace(-5,5,100)
y = x**2
plt.plot(x,y , 'r--')
plt.xlabel('x')
plt.ylabel('$y = x^2$')
plt.show()
````

````python
#You can also use rgb codes to define your color, and change the linewidth 
x = np.linspace(-5,5,100)
y = x**2
plt.plot(x,y , '--', color = ([0.8,0.1,0.25]), lw = 3)
plt.xlabel('x')
plt.ylabel('$y = x^2$')
plt.show()
````

````python
#you can also plot multiple lines on the same canvas
x = np.linspace(-5,5,20)
plt.plot(x, x**2, 'ko')
plt.plot(x, x**3, 'r*')
plt.show()
````

````python
#and change the axis limits to better represent your data
x = np.linspace(-5,5,20)
plt.plot(x, x**2, 'ko')
plt.plot(x, x**3, 'r*')
plt.ylim([-50,50])
plt.title(f'Plot of Various Polynomials from {x[0]} to {x[-1]}')
plt.xlabel('X axis', fontsize = 18)
plt.ylabel('Y axis', fontsize = 18)
plt.show()
````

````python
#and add legends for clarification 
x = np.linspace(-5,5,20)
plt.plot(x, x**2, 'ko', label='$y = x^2$')
plt.plot(x, x**3, 'r*', label='$y = x^3$')
plt.ylim([-50,50])
plt.title(f'Plot of Various Polynomials from {x[0]} to {x[-1]}')
plt.xlabel('X axis', fontsize = 18)
plt.ylabel('Y axis', fontsize = 18)
plt.legend(fontsize=18)
plt.show()
````

````python
#you can also create subplots by defining the number of row and columns 
x = np.arange(11)
y = x**2

plt.figure(figsize = (14, 8))

plt.subplot(2, 3, 1)
plt.plot(x,y)
plt.title('Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.subplot(2, 3, 2)
plt.scatter(x,y)
plt.title('Scatter')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.subplot(2, 3, 3)
plt.bar(x,y)
plt.title('Bar')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.subplot(2, 3, 4)
plt.loglog(x,y)
plt.title('Loglog')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(which='both')

plt.subplot(2, 3, 5)
plt.semilogx(x,y)
plt.title('Semilogx')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(which='both')

plt.subplot(2, 3, 6)
plt.semilogy(x,y)
plt.title('Semilogy')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.tight_layout()

plt.show()
````

## 3d Plots

you can also use Matplotlib to create beautiful 3d plots

````python
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
z = np.arange(0, 10*np.pi, np.pi/50)
x = np.sin(z)
y = np.cos(z)

ax.plot3D(x, y, z)
ax.set_title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

plt.show()
````

````python
x = np.random.random(50)
y = np.random.random(50)
z = np.random.random(50)

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.grid()

ax.scatter(x, y, z, c = 'r', s = 50)
ax.set_title('3D Scatter Plot')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

plt.show()
````

````python
fig = plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')

x = np.arange(-5, 5.1, 0.2)
y = np.arange(-5, 5.1, 0.2)

X, Y = np.meshgrid(x, y)
Z = np.sin(X)*np.cos(Y)

surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

fig.colorbar(surf, shrink=0.5, aspect=8)

plt.show()
````

## Animations 

Matplotlib also allows you to create animations using loops 

you need to run the command


before attempting this code

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as manimation
````

````python
n = 1000
x = np.linspace(0, 6*np.pi, n)
y = np.sin(x)

# Define the meta data for the movie
FFMpegWriter = manimation.writers['ffmpeg']
metadata = dict(title='Movie Test', artist='Matplotlib',
                comment='a red circle following a blue sine wave')
writer = FFMpegWriter(fps=15, metadata=metadata)

# Initialize the movie
fig = plt.figure()

# plot the sine wave line
sine_line, = plt.plot(x, y, 'b')
red_circle, = plt.plot([], [], 'ro', markersize = 10)
plt.xlabel('x')
plt.ylabel('sin(x)')

# Update the frames for the movie
with writer.saving(fig, "writer_test.mp4", 100):
    for i in range(n):
        x0 = x[i]
        y0 = y[i]
        red_circle.set_data(x0, y0)
        writer.grab_frame()
````
